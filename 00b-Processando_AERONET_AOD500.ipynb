```python
var = 'AOD_500nm'
mins = np.arange(0,1440,1)
aodmeses = []

for dss in datasets:
    l = list(dss['Date(dd:mm:yyyy)'].values)
    dias = []
    for j in range(len(l)):
        dia = l[j].split(':')[0]
        mes = l[j].split(':')[1]
        ano = l[j].split(':')[2]
        day = dia+'/'+mes+'/'+ano
        dias.append(day)
    days1 = sorted(list(set(dias)))
    days2 = sorted(list(set(l)))
    dfsdias = []
    
    for d in range(len(days2)):
        df = dss.loc[dss['Date(dd:mm:yyyy)'] == days2[d]]
        dfsdias.append(df)

    aodsdias = []
    for d in dfsdias:
        falta = len(mins)-len(d[var])
        aod = list(d[var])
        for f in range(falta):
            aod.append(np.nan)
        aodsdias.append(aod)
    
    dic = {}
    for i in range(len(days1)):
        dic[days1[i]] = aodsdias[i]
        
    aod = pd.DataFrame(dic)
    aodmeses.append(aod)
```
