```python
### Guardando os subtítulos de cada voo (no formato "dd/mm/aaa starting at hh:mm:ss") para inserir na figura
### final:
subtitulos = []
for j in range(len(arquivos)):
    t1 = arquivos[j].split('/')[-1].split('.')[0].split('_')[2]
    mes = t1[4:6]
    dia = t1[6:8]
    hora = t1[8:10]
    minuto = t1[10:12]
    seg = t1[12:]
    data_american = mes+'/'+dia+'/2014 - '+hora+':'+minuto+':'+seg
    data_british = dia+'/'+mes+'/2014 - '+hora+':'+minuto+':'+seg
#     subtitulos.append(data_american)
    subtitulos.append(data_british)
    
### Inserindo o subtitulo do dia 03/03/2014 na lista: 
# subtitulos.insert(4, '03/03/2014 starting at '+ds04['Time'][0])
subtitulos.insert(4, '03/03/2014 - 17:45:40')
subtitulos
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import warnings
import matplotlib.patches as patches
warnings.filterwarnings("ignore")

#### Fazendo a figura e os eixos (os "axs"):
fig, axs = plt.subplots(4, 4, figsize=(16, 16), facecolor='white', sharey=True, sharex=True)
axes = [axs[0][0], axs[0][1], axs[0][2], axs[0][3],
        axs[1][0], axs[1][1], axs[1][2], axs[1][3],
        axs[2][0], axs[2][1], axs[2][2], axs[2][3],
        axs[3][0], axs[3][1], axs[3][2], axs[3][3]]

### Setando o tamanho da fonte e cores:
ms = 14
tf = 13
limsx = [-61.5, -59.0]
limsy = [-4, -2.4]
cores=['black', 'red', 'purple']

for i in range(len(axes)):
    #### Personalizando os eixos x e y:
    plt.setp(axes[i].get_xticklabels(), rotation=0, horizontalalignment='center', fontsize=tf, color='black')
    plt.setp(axes[i].get_yticklabels(), rotation=0, horizontalalignment='right', fontsize=tf, color='black')
    
#     if i==3 or i==4 or i==5:
    if i==3 or i==5:
        axes[i].set_title(subtitulos[i], fontsize=tf, color=cores[1])
    elif i==4:
        axes[i].set_title(subtitulos[i], fontsize=tf, color=cores[2])
    else:
        axes[i].set_title(subtitulos[i], fontsize=tf, color=cores[0])
    if i==0 or i==4 or i==8:
        axes[i].set_ylabel('Latitude (°)', fontsize=tf, color='black')
    if i==12:
        axes[i].set_ylabel('Latitude (°)', fontsize=tf, color='black')
        axes[i].set_xlabel('Longitude (°)', fontsize=tf, color='black')
    if i==13 or i==14 or i==15:
        axes[i].set_xlabel('Longitude (°)', fontsize=tf, color='black')
        
    axes[i].plot(voos_iop1[i]['Lon'], voos_iop1[i]['Lat'])
    axes[i].set(xlim=limsx, ylim=limsy, facecolor='white')
    axes[i].grid(b='True', which='both', axis='both', color='gray', linestyle='--', linewidth=0.35)
    axes[i].plot(-60, -3.1, marker='x', color="black", markersize=ms)
    axes[i].plot(-60.5987, -3.2133, marker='o', mfc='none', color="red", markersize=ms)
    axes[i].yaxis.set_major_locator(ticker.MultipleLocator(0.2))
    axes[i].xaxis.set_major_locator(ticker.MultipleLocator(0.5))
#     axes[i].add_patch(patches.Rectangle((-61.0,-3.4),0.55,0.4,edgecolor='limegreen',fill=False,ls='--',lw=3))
#     axes[i].add_patch(patches.Rectangle((-60.65,-3.25),0.15,0.1,edgecolor='limegreen',fill=False,ls='--',lw=3))
    axes[i].add_patch(patches.Rectangle((-60.75,-3.3),0.3,0.2,edgecolor='limegreen',fill=False,ls='--',lw=3.5))
    
plt.tight_layout()
plt.savefig('/mnt/HD_500GB_WD/historia2/Z-Copia_Notebooks_selecionados-Final/Trajetorias_IOP1.png', 
            dpi=400, bbox_inches='tight', facecolor='white')
```
