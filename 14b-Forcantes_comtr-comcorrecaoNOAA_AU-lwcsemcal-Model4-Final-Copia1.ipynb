```python
for i, d in enumerate(list(dfp.columns)):
    dia = d.split('/')[0]
    mes = d.split('/')[1]
    if d == '05/10/2014':
        print('Primeiro período: 0 a {0}'.format(i))
    elif d == '28/10/2014':
        print('Segundo período: {0} a {1}'.format(i, len(dfp.T)))

diasp1, diasp2 = [], []
for i in range(len(dfp.columns)):
    if i <= 38:
        diasp1.append(dfp.columns[i])
    elif i >= 39:
        diasp2.append(dfp.columns[i])
```

```python
################## Fazendo o merge dos dataframes para 2014 (limpo+trans2+poluido1+trans3+poluido2+trans4):
merge1 = dfl.merge(dft2, left_index=True, right_index=True)
merge2 = merge1.merge(dfp[diasp1], left_index=True, right_index=True)
merge3 = merge2.merge(dft3, left_index=True, right_index=True)
merge4 = merge3.merge(dfp[diasp2], left_index=True, right_index=True)
merge5 = merge4.merge(dft4, left_index=True, right_index=True)
merge5
```

```python
days14 = pd.date_range(start="2014-01-01", end='2014-12-31',freq='D').strftime('%d/%m/%Y')
df_nans = pd.DataFrame(columns=days14, index=['sza', 'irrad'])
booleanos = [dia in list(merge5.columns) for dia in list(df_nans.columns)]
for i in range(len(booleanos)):
    if booleanos[i] == True:     # Há dados neste dia
        pass
    elif booleanos[i] == False:  # Não há dados neste dia
        merge5.insert(i, column=df_nans.columns[i], value=df_nans.T.iloc[i].values)
        
merge5
```

```python
path = '/mnt/HD_500GB_WD/historia2/ZZ-TabelasNOAA/'

arqs = sorted(glob.glob(path+'NOAA_???.ods'))
tabelas14 = []
for i in range(len(arqs)):
#     print(arqs[i])
    df = read_ods(arqs[i], 1, headers=True)
    tabn = df[['Date', 'Sun Rad Vector (AUs)', 'Solar Noon (LST)', 'Solar Zenith Angle (deg)']]
    tabelas14.append(tabn)
    
szas12LTnoaa, d12LT = [], []
for t in tabelas14:
    szas12LTnoaa.append(t.iloc[119][3])
    d12LT.append(t.iloc[119][1])
```

```python
d2s, razoes, reflets = [], [], []
for i in range(len(merge5.loc['irrad'])):
    E = merge5.loc['irrad'][i]
    sza = merge5.loc['sza'][i]*(np.pi/180)
    mu0 = np.cos(sza)
    d2 = d12LT[i]**2 
    reflet = (E*d2)/mu0
    d2s.append(d2)
    razoes.append(mu0/d2)
    reflets.append(reflet)
```

```python
fig, ax = plt.subplots(figsize=(8, 6), facecolor='white')
tf = 14
limitesx = [0, 370]
limitesy = [0.98, 1.02]
plt.setp(ax.get_xticklabels(), rotation=0, horizontalalignment='center', fontsize=tf, color='black')
plt.setp(ax.get_yticklabels(), rotation=0, horizontalalignment='right', fontsize=tf, color='black')
ax.set_xlabel('Tempo (dias julianos)', fontsize=tf)
ax.set_ylabel('d (UA)', fontsize=tf)
ax.grid(axis='both', linestyle='--', linewidth=0.22, color='grey')
ax.set(xlim=limitesx, ylim=limitesy, facecolor='white')
ax.xaxis.set_major_locator(ticker.MultipleLocator(30))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.005))

ax.scatter(np.arange(len(merge5.columns)), np.sqrt(d2s), s=2.5, color='green')

# plt.savefig('/mnt/HD_500GB_WD/historia2/Z-Copia_Notebooks_selecionados-Final/d2_NOAA_2014.png', 
#             dpi=400, bbox_inches='tight', facecolor='white')
```
