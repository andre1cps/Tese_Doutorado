```python
### Pacotes para abrir arquivos em pastas:
import os
import glob
import warnings
warnings.filterwarnings('ignore')

### Pacotes para arrumar os dados:
import xarray as xr
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
### Verificando a pasta com os arquivos AERO_MERGED.nc dos IOPs:
pasta_selecao = '/media/andre/SAMSUNG/03_HD_500GB_WD/G1_Micael/selecionados/iopsjuntos'
arquivosaero = sorted(glob.glob(pasta_selecao+'/AERMERGE_G1_2014??????????_R2_L1_goamazon001s.nc'))
for j in range(len(arquivosaero)):
    t = arquivosaero[j].split('/')[-1]
    print('Arq. {0}: {1}'.format(j,t))
    
### Guardando as strings com os dias de cada .nc numa lista e com os voos em outra lista:
voosaero = []
diasaero = []
for j in range(len(arquivosaero)):
    v = arquivosaero[j].split('/')[-1].split('.')[0].split('_')[2]
    d = v[4:8]
    voosaero.append(v)
    diasaero.append(d)
    
### Para todos os AERO.nc, guardando os xarray.dataset na lista 'dsaero':
dsaero = []
for i in range(len(arquivosaero)):
    ds = xr.open_dataset(pasta_selecao+'/AERMERGE_G1_'+voosaero[i]+'_R2_L1_goamazon001s.nc',
                         decode_times=False, mask_and_scale=True)
    dsaero.append(ds)
```

```python
### Abrindo os arquivos .txt (dos dois IOP) da pasta 'comstock-wcm', que contém dados de TWC e LWC:
pasta = '/media/andre/SAMSUNG/03_HD_500GB_WD/G1_Micael/Dados_G1_Completos/comstock-wcm/selecionados/iopsjuntos/'
arquivos = sorted(glob.glob(pasta+'aaf.wcm001s.g1.goamazon.2014????a.txt'))
for j in range(len(arquivos)):
    t = arquivos[j].split('/')[-1]
    print('Arq. {0}: {1}'.format(j,t))

### Para cada um dos .txt, excluindo a coluna 'time(UTC)':
txts = []
for file in arquivos:
    csv = pd.read_csv(file, sep=",", header=0, na_values=["-9999.00"])
    csv.drop(['Time(UTC)'], axis=1, inplace=True)
    txts.append(csv)
```

```python
### Adicionando os dados de TWC e LWC aos xarray.Datasets que já existiam:
for i, ds in enumerate(dsaero):
    TWC = np.array(txts[i]['WCM_TWC(g/m3)'])
    LWC_083 = np.array(txts[i]['WCM_083(g/m3)'])
    LWC_021 = np.array(txts[i]['WCM_021(g/m3)'])
    ds["TWC"]=(['POINTS'],TWC)
    ds["LWC_083"]=(['POINTS'],LWC_083)
    ds["LWC_021"]=(['POINTS'],LWC_021)
```

```python
### Para todos os dsaero, convertendo os xarray para pd.Dataframe e guardando os dataframes na lista 'dfaero':
dfaero = []
for ds in dsaero:
    df = ds.to_dataframe()
    dfaero.append(df)
```

```python
### Listando os arquivos FCDP.nc, os quais contém os dados de concentração numérica de gotas de nuvem:
pasta_fcdp = '/media/andre/SAMSUNG/03_HD_500GB_WD/G1_Micael/Dados_G1_Completos/mei-fcdp_nc/'
pasta_selecao = pasta_fcdp+'selecionados/'
arquivos = sorted(glob.glob(pasta_selecao+'FCDP_G1_2014??????????_R2_GoAmazon001s.nc'))
for j in range(len(arquivos)):
    t = arquivos[j].split('/')[-1]
    print('Arq. {0}: {1}'.format(j,t))
    
### Guardando as strings com os dias de cada .nc numa lista e com os voos em outra lista:
voos = []
dias = []
for j in range(len(arquivos)):
    v = arquivos[j].split('/')[-1].split('.')[0].split('_')[2]
    d = v[4:8]
    voos.append(v)
    dias.append(d)
    
### Para todos os FCDP.nc, guardando os xarray.dataset na lista 'dsfcdp' e os dataframes na lista 'dffcdp':
dffcdp = []
for i in range(len(arquivos)):
    ds = xr.open_dataset(pasta_selecao+'FCDP_G1_'+voos[i]+'_R2_GoAmazon001s.nc', decode_times=False, 
                         mask_and_scale=True)
    df = ds.to_dataframe()
    dffcdp.append(df)
```

```python
### Como há menos medidas do FCDP do que do AERO, pegando somente as linhas do AERO nas quais
### há medidas do FCDP (somente para os dataframes):
dfaerovalid = []
for i in range(len(dfaero)):
    dfaerovalid.append(dfaero[i].iloc[:len(dffcdp[i])])
    
### Verificando se as linhas de cada par (dffcdp,dfaerovalid) correspondem entre si:
tamf = []
for i in range(len(dffcdp)):
    tamf.append(len(dffcdp[i]))
tama = []
for i in range(len(dfaerovalid)):
    tama.append(len(dfaerovalid[i]))
if(set(tamf) == set(tama)):
    print("Lists are equal")
```

```python
dfs_juntos = []
for i in range(len(dfaerovalid)):
    dffcdp[i].drop(['Second'], axis=1, inplace=True)
    merge = dfaerovalid[i].merge(dffcdp[i], left_index=True, right_index=True)
    colunas_a_excluir = list(merge.columns)[4:17]+list(merge.columns)[18:20]+list(merge.columns)[21:23]+list(merge.columns)[24:41]
    merge.drop(colunas_a_excluir, axis=1, inplace=True)
    dfs_juntos.append(merge)
```

```python
datasets = []
for df in dfs_juntos:
    ds = xr.Dataset.from_dataframe(df)
    datasets.append(ds)
```

```python
### Função que, dado um Dataset do Xarray e uma variavel dentro do Dataset, retorna uma lista de tamanho igual a
### 86400, com as medidas já posicionadas no ponto temporal (segundos) correto e o resto dos pontos preenchidos
### com NaN. Para fazer isso a função primeiro identifica em qual ponto temporal a primeira medida foi feita, e
### com isso é contruída uma lista de cheia de NaN. Depois essa lista é concatenada com a lista das medidas, e
### por fim o resultado dessa concatenação é concatenado com uma segunda lista de NaN, de modo que a lista final
### resultante tenha um tamanho igual ao número de segundos em 1 dia. Tudo isso é feito pois as medidas do avião
### foram tomadas de segundo em segundo. 
def gera_dados(xarrayds, var):
    secondsday = 60*60*24
    n1 = []
    for p1 in range(int(xarrayds['Start_UTC'].values[0])-1):
        n1.append(np.nan)
    f = n1+list(xarrayds[var].values)
    n2 = []
    for p2 in range(secondsday-len(f)):
        n2.append(np.nan)
    dados = f+n2
    return dados
```

```python
### Separando as variáveis de interesse e aplicando a função "gera_dados" nelas. Isso é feito para cada um dos
### .nc, de modo que no final a lista "lista" tenha 31 sublistas:
variaveis = list(dfs_juntos[0].columns)[1:]

listas = []
for d in range(len(datasets)):
    listas.append([gera_dados(datasets[d], v) for v in variaveis])
```

```python
### Para cada um dos .nc construindo uma lista de tuplas. Cada tupla contém 3 informações:
### (Index com strings no formato 'hh:mm:ss', DatetimeIndex com strings o formato 'AAAA-MM-DD hh:mm:ss', string
### no formato 'AAAA:MM-DD'):
secondsday = 60*60*24
indexes = []
for d in range(len(datasets)):
    diads = '2014'+'-'+dsaero[d].attrs['SDATE'].split(', ')[1]+'-'+dsaero[d].attrs['SDATE'].split(', ')[2]
    indexes.append((pd.date_range(start=diads, freq='1s', periods=secondsday).strftime('%H:%M:%S'),
                    pd.date_range(start=diads, freq='1s', periods=secondsday),
                    diads))
```

```python
### Para cada um dos .nc criando um Xarray Dataset, e depois guardando cada um na lista "xarraysDatasets_ii".
### Os Xarray Datasets são criados a partir de um dicionário com as variáveis de interesse. Esse dicionário tem
### que estar no formato {nome_variavel: (dims,dados,attrs)}, onde o 'nome_variavel' é uma string, 'dims' é uma
### lista de strings, 'dados' é uma lista com os dados e 'attrs' é um dicionário com os atributos (unidade, 
### nome extenso da variável, etc), todos strings. Na criação dos Xarrays Datasets também são usados os 
### dicionários "coords" e "attrs" (esse último com os atributos globais do xarray Dataset):
xarraysDatasets_ii = []
for d in range(len(datasets)):
    data_vars = {}
    for i in range(len(variaveis)):
        data_vars.update({variaveis[i]: (['time'],listas[d][i])})
    coords = {'time': (['time'],indexes[d][1])}
    attrs = {'creation_date':datetime.now(), 'author':'Andre Pugliesi', 'dia':indexes[d][2]}
    dss = xr.Dataset(data_vars=data_vars, coords=coords, attrs=attrs)
    xarraysDatasets_ii.append(dss)
    
xarraysDatasets_ii[0]
```

```python
medias_5s = []
for d in range(len(datasets)):
    rawdata1 = xarraysDatasets_ii[d].resample(time="5s").mean(skipna=True)
    medias_5s.append(rawdata1)
```

```python
indexes2 = []
for d in range(len(datasets)):
    diads = '2014'+'-'+dsaero[d].attrs['SDATE'].split(', ')[1]+'-'+dsaero[d].attrs['SDATE'].split(', ')[2]
    indexes2.append((pd.date_range(start=diads, freq='5s', periods=17280).strftime('%H:%M:%S'),
                     pd.date_range(start=diads, freq='5s', periods=17280), diads))
```

```python
### Para cada um dos .nc criando um Xarray Dataset, e depois guardando cada um deles na lista "xarraysDatasets".
### Os Xarray Datasets são criados a partir de um dicionário com as variáveis de interesse. Esse dicionário tem
### que estar no formato {nome_variavel: (dims,dados,attrs)}, onde o 'nome_variavel' é uma string, 'dims' é uma
### lista de strings, 'dados' é uma lista com os dados e 'attrs' é um dicionário com os atributos (unidade, 
### nome extenso da variável, etc), todos strings. Na criação dos Xarrays Datasets também são usados os 
### dicionários "coords" e "attrs" (esse último com os atributos globais do xarray Dataset):
xarraysDatasets = []
for d in range(len(medias_5s)):
    data_vars = {'UTC': (['time'],indexes2[d][0],{'units':'hh:mm:ss'}),
                 'pts_tempo': (['time'],list(range(17280)),{'units':'unitless'})}
    for i in range(len(variaveis)):
        data_vars.update({variaveis[i]: (['time'],medias_5s[d][variaveis[i]].values)})
    coords = {'time': (['time'],indexes2[d][1])}
    attrs = {'creation_date':datetime.now(), 'author':'Andre Pugliesi', 'dia':indexes2[d][2]}
    dss = xr.Dataset(data_vars=data_vars, coords=coords, attrs=attrs)
    xarraysDatasets.append(dss)
    
xarraysDatasets[0]
```

```python
dfs5s = []
for ds in xarraysDatasets:
    df = ds.to_dataframe()
    dfs5s.append(df)
```

```python
### Para cada um dos dataframes na lista 'dfs' substituindo os valores ausentes (-9.99) por np.nan:
for i in range(len(dfs5s)):
    dfs5s[i].mask(dfs5s[i] == -9.99, np.nan, inplace=True)
```
